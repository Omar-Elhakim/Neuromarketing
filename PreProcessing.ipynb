## Read the dataset files

```python
dataset= np.load('SEED/DatasetCaricatoNoImage.npz')['arr_0']
labels= np.load('SEED/LabelsNoImage.npz')['arr_0']
subjects= np.load('SEED/SubjectsNoImage.npz')['arr_0']
```

## The dataset has the following shape: 

```python
dataset.shape
```

### This can be interpreted as (window , frequency band, channel). This means that for every window , we have its DE value for each frequency band across every channel.

The frequency bands used:

* Delta (1–4 Hz)

* Theta (4–8 Hz)

* Alpha (8–14 Hz)

* Beta (14–31 Hz)

* Gamma (31–50 Hz)

The labels are mapped like this 

0 → negative  
1 → neutral  
2 → positive

```python
labels.shape,np.unique(labels)
```
